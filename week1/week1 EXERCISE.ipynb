To demonstrate your familiarity with OpenAI API, and also Ollama, build a tool that takes a technical question,  
and responds with an explanation. This is a tool that you will be able to use yourself during the course!

````python
# imports
import os
from openai import OpenAI
from IPython.display import display, Markdown, update_display
from dotenv import load_dotenv
import ollama
load_dotenv(override=True)
````

````python
# constants

MODEL = 'gemini-2.5-flash'
MODEL_LLAMA = 'llama3.2'
api_key = os.getenv("GEMINI_API_KEY")

gemini = OpenAI(
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/",
    api_key=api_key
)
````

````python
# here is the question; type over this to ask something new

question = """
What is Python? Is it a snake or something else?
"""
````

````python
# prompts
system_prompt = """  
You are a helpful and funny assistant. If I ask you questions about maths, data science, or Python, you should answer sarcastically.
"""

user_prompt = 'please give a detailed answer to the following question:' + question
````

````python
messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
]
````

````python
stream = gemini.chat.completions.create(
    model=MODEL,
    messages=messages,
    stream=True
)

response = ""
display_handle = display(Markdown(""), display_id=True)
for chunk in stream:
    response += chunk.choices[0].delta.content or ''
    response = response.replace("```", "").replace("markdown", "")
    update_display(Markdown(response), display_id=display_handle.display_id)
````

````python
response = ollama.chat(model=MODEL_LLAMA, messages=messages)
reply = response['message'] ['content']
display(Markdown(reply))
````
